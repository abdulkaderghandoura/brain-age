```python
# Import necessary modules
import os
import numpy as np
import mne
import pickle
from tqdm import tqdm
from scipy import signal
```

```python
# import sys
# sys.path.append('../../src/preprocessing')
from preprocessing_parallel import preprocess_file

file_path = '/data0/practical-sose23/brain-age/data/bap/healthy_controls/preprocessed/Exp_eyes_closed_vpH38_eyes_closed.vhdr'
sfreq = 300
references = ['average']
filters = {"lpf":100, "hpf":1, "nf":50}

eeg_obj1, _ = preprocess_file(file_path, sfreq, references, filters)

eeg_obj2, _ = preprocess_file(file_path, None, references, filters)

eeg_obj3, _ = preprocess_file(file_path, sfreq, [], filters)

eeg_obj4, _ = preprocess_file(file_path, None, [], filters)
```

```python
# Load the standard 10-20 montage (positions of electrodes)
montage = mne.channels.make_standard_montage('standard_1020')

# Set the montage for the raw object
eeg_obj1.set_montage(montage)

eeg_obj1.compute_psd(fmax=110).plot()

# With downsampling and with referencing
```

```python
# Set the montage for the raw object
eeg_obj2.set_montage(montage)

eeg_obj2.compute_psd(fmax=110).plot()

# Without downsampling and with referencing
```

```python
# Set the montage for the raw object
eeg_obj3.set_montage(montage)

eeg_obj3.compute_psd(fmax=110).plot()

# With downsampling and without referencing
```

```python
# Set the montage for the raw object
eeg_obj4.set_montage(montage)

eeg_obj4.compute_psd(fmax=110).plot()

# Without downsampling and without
```

### Set-up functions for parallel extraction

```python
def mat_files(members, substring):
    for tarinfo in members:
        if substring in tarinfo.name:
            yield tarinfo
            
def extract_files(source_tar, dest_dir, type_files, target_substring):
    try:
        tar = tarfile.open(source_tar)
        tar.extractall(path=dest_dir, members=mat_files(tar, target_substring))
    except EOFError:
        pass

extract_files = partial(extract_files, dest_dir=dest_dir, type_files=mat_files, target_substring=target_substring)
```

### Extract files with 128 cpus

```python
Parallel(n_jobs=128)(delayed(extract_files)(f) for f in tar_files)
```
